```python
everyColumn = ['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol','quality']
# Replace all null values 
```

```python
for column in everyColumn:
    data[column].fillna(data[column].median(), inplace=True)
```

```python
#Replacing null values with the median of the column is a common technique used in data analysis when missing data is present.
# when we leave these null places unattended to, it leads to drawing of inaccurate inferences about data.
# We could use median, mode or mean but inputting with the median is more robust than inputting with the mean, because it mitigates the effect of outliers.
```
